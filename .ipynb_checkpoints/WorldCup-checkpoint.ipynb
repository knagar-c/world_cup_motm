```python
def train(x, y, **kwargs):
    alphas = kwargs.get('alphas', [1, 0.1, 0.001, 0.0001])
    mtype = kwargs.get('model_type')
    model = LogisticRegression()

    # Train model
    model.fit(x, y)
    
    return model
```

```python
def predict_and_score(model, x, y):
    predictions = model.predict(x)
    rmse = np.sqrt(mean_squared_error(predictions, y))
    return [predictions, rmse]
```

```python
x_train, x_test, y_train, y_test = model_selection.train_test_split(train_df, y.values, test_size=0.30, random_state=10)
```

```python
rf = RandomForestClassifier()
```

```python
all_vars = x_train.columns.tolist()
top_vars = ['Attempts', 'Goal Scored', 'Ball Possession', 'Passes', 'Free Kicks', 'On-Target', 'Saves', 'Yellow Card', 'Pass Accuracy %', 'Off-target', 'Blocked', 'Distance Covered (Kms)']
bottom_vars = [cols for cols in all_vars if cols not in top_vars]
x_train    = x_train.drop(bottom_vars, axis=1)  
x_test     = x_test.drop(bottom_vars, axis=1)  
```

```python
%%time

exp = cortex.experiment('kaggle/worldcup')
exp.set_meta('style', 'supervised')
exp.set_meta('function', 'regression')

logrmse = 1.0
logmodel_type = None
logmodel = None

with exp.start_run() as run:
    alphas = [1, 0.1, 0.001, 0.0005]
    print('---'*30)
    print('Training model using logistic regression algorithm')
    model_type = 'Logistic'
    model = train(x_train, y_train, model_type=model_type, alphas=alphas)
    [predictions, rmse] = predict_and_score(model, x_train, y_train)
    print('Training error:', rmse)
    [predictions, rmse] = predict_and_score(model, x_test, y_test)
    print('Testing error:', rmse)
    
    logrmse = rmse
    logmodel = model
    logmodel_type = model_type
    
    r2 = model.score(x_test, y_test)
    run.log_metric('r2', r2)
    run.log_metric('rmse', logrmse)
    run.log_param('model_type', logmodel_type)
    run.log_param('alphas', alphas)
    run.log_artifact('model', logmodel)

print('---'*30)
```

```python
print('Model: ' + logmodel_type)
print('Testing error: %.6f' % logrmse)
print('R2 score: %.6f' % r2)
```

```python
def fix_columns(pipeline, df):
    all_cols = pipeline.get_context('columns')
    missing_cols = set(all_cols) - set(df.columns)
    for c in missing_cols:
        df[c] = 0
    
    # make sure we have all the columns we need
    assert(set(all_cols) - set(df.columns) == set())
    
    return df[all_cols]
```
